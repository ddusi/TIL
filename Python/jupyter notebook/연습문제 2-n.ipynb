# 연습문제 2-5

```python
scoreList = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]
total = 0
# 코드 작성 - List pop() 함수 이용
for i in reversed(scoreList):
    score = scoreList.pop()
    
    if score >= 50:
        total += score

print(total)
```

# 질문 사항

```python
scoreList = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]
total = 0
# 코드 작성 - List pop() 함수 이용
for i in scoreList:
    score = scoreList.pop(0)
    print(score)
    
    if score >= 50:
        total += score

print(total)
```

```python
scoreList = [33, 55, 67, 82, 45, 33, 90, 87, 100, 25]
total = 0

while scoreList:
    score = scoreList.pop()

    if score >= 50:
        total += score

print(total)
```

```python
studentInfo = {
 'a':33, 'b':55, 'c':67, 'd':82, 'e':45,
 'f':33, 'g':90, 'h':87, 'i': 100, 'j':25
}
total = 0
# 코드 작성 
key1 = list(studentInfo.keys()) #딕셔너리의 키값만 뽑아서 리스트화 시켰다.

for i in key1: #그 리스트값을 for문으로 차례로 뽑아 아래와 같은 문장에 대입
    a = studentInfo.get(i) #딕셔너리 메서드인 get을 이용하여 values의 값에 접근.
    if a >= 80:
        total += a

print(total)
```

```python
studentInfo = {
 'a':33, 'b':55, 'c':67, 'd':82, 'e':45,
 'f':33, 'g':90, 'h':87, 'i': 100, 'j':25
}
total = 0
# 코드 작성 - Dictionary values(), List pop() 등 이용
val = list(studentInfo.values())

while val:
    val2 = val.pop()
    
    if val2 >= 80:
        total += val2
    

print(total)
```

# 연습문제 2-6 

```python
import random # random 모듈 import

# 코드 작성
while True:
    a = random.randint(1,6)
    b = random.randint(1,6)
    print (a,'/',b)
    if a == b:
        print('두 눈이 같으므로 종료')
        break

# random.randint(1, 6) 함수 이용
# 두 눈이 같을 때 break
```

```python
import random
lottoList = list() # 빈 List 생성
# 코드 작성
for i in range(6):
    a = random.randint(1, 45)
    lottoList.append(a)

# 반복문 조건으로 len() 함수 이용
# random.randint(1, 45) 함수 이용
# 번호 생성 후 lottoList에 append
print(lottoList)
```

```python
import random
lottoList = list()

while True:
    a = random.randint(1, 45)
    lottoList.append(a)
    if len(lottoList) >= 6:
        break

print(lottoList)
```
