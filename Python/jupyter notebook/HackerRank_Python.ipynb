### 변수 X, Y, Z를 선언한다. 이때, 각각에 인덱스 i,j,k를 사용하여 리스트[i, j, k]의 모든 경우의 수를 출력한다.
### (단, i+j+k = n 인 것은 제외한다.)

```python
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
```

```python
result = []
for i in range(x+1):
    for j in range(y+1):
        for k in range(z+1):
            print(k)
            if n != i+j+k:
                temp = [i,j,k]
                result.append(temp)

print(result)
```
