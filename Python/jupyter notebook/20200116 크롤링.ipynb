# 연습문제 10-7 인터파크

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium import webdriver as wd


driver = wd.Chrome(executable_path='chromedriver.exe')
driver.get('http://tour.interpark.com/')

driver.find_element_by_id('SearchGNBText').send_keys('달랏')
driver.find_element_by_class_name('search-btn').click()

driver.implicitly_wait (10)

# 해외여행 목록을 간략하게 표시한 페이지
driver.find_element_by_css_selector('.panelZone > .oTravelBox .moreBtn').click()

import time
# 해외여행 상세 목록 페이지(실제 작업할 페이지 1~6)
time.sleep(2)
# 1. 해외여행 패키지 개수확인
totalCnt = driver.find_element_by_id('totalAllCnt')
print(totalCnt.text)
totalCnt = int(totalCnt.text)

# 2. 전체 페이지 수를 확인
totalPage = totalCnt // 10 + 1
if totalCnt % 10 == 0:
    totalPage -= 1
print(totalPage)

for page in range(1, totalPage + 1):
    driver.execute_script('searchModule.SetCategoryList(%d, '')'% page)
    # 1 페이지로 이동
    time.sleep(2)
    boxItems = driver.find_elements_by_css_selector('.oTravelBox .boxItem')
    for item in boxItems:
        print(item.find_element_by_class_name('proTit').text)
        print(item.find_element_by_css_selector('img').get_attribute('src'))
```
