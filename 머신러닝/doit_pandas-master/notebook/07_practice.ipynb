```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.columns)
```

```python
ebola.iloc[:5,[0,1,2,3,10,11]]
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long.head()

# id_vars는 기준이될, 살리고싶은 열을 쓰는 것. 
```

```python
# split메서드로 열이름 분리하기. 
variable_split = ebola_long.variable.str.split('_')
print(variable_split[:5])
# ebola_long에 있는 테이블에서 variable이름의 열안에 str(문자열)에 split()메서드 적용.
```

variable_split에 저장된 값의 자료형은 시리즈이고 각각의 시리즈에 저장된 값의 자료형은 리스트.

```python
type(variable_split)
```

```python
type(variable_split[0])
# 시리즈 안에 list로 저장됨.
```

```python
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
```

```python
status_values.tail()
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
```

```python
ebola_long.head()
```

# 이보다 좋은 기능인 concat메서드가 있다. 
이 메서드를 이용하면, split메서드로 분리한 데이터를 바로 데이터프레임에 추가할 수 이싿. 


```python
variable_split = ebola_long.variable.str.split('_',expand=True)
variable_split.columns = ['status','country']
```

```python
pd.concat([ebola_long, variable_split], axis=1)
```

# 기상 데이터의 여러 열을 하나로 정리하기

```python
weather = pd.read_csv('../data/weather.csv') 
print(weather.iloc[:5, :11])
```

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], 
    columns='element', 
    values='temp'
)

print(weather_tidy)
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```

```python
weather_tidy.reset_index()
# pivot_table() 메서드는 좀더 아름답게 이쁘게 열을 분리해주는 메서드
```

# 뉴욕 택시 데이터 준비

```python
import os 
import urllib.request

# 네트워크 상태에 따라 5 ~ 15분이 소요됩니다.
with open('../data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line == 5:
            break 
        fn = url.split('/')[-1].strip()
        fp = os.path.join('', '../data', fn)
        print(url)
        print(fp)
        urllib.request.urlretrieve(url, fp)
```
